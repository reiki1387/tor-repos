```python
import pandas as pd

# Load the CSV
# Replace 'yourfile.csv' with the actual path to your file
df = pd.read_csv(r'C:\Users\FRED\Desktop\MSE Capstone Project Files\Historical Forex Data\EURUSD_M15(raw).csv')
df
```

```python
#drop columns
df = df.drop(columns = ["Unnamed: 1", "Unnamed: 2" , "Unnamed: 3", "Unnamed: 4", "Unnamed: 5", "Unnamed: 6", "Unnamed: 7"] )
df
```

```python

# Split the single column by tab characters
df_split = df['TimeOpenHighLowCloseVolume'].str.split('\t', expand=True)
df_split
```

```python
# Assign proper column names
df_split.columns = ['Datetime', 'Open', 'High', 'Low', 'Close', 'Volume1', 'Volume2']
df_split
```

```python
# Split 'Datetime' into Date and Time
df_split[['Date', 'Time']] = df_split['Datetime'].str.split(' ', expand=True)
df_split
```

```python
#Convert to float
float_cols = ['Open', 'High', 'Low', 'Close']
df_split[float_cols] = df_split[float_cols].astype(float)
```

```python
# Combine the two volume fields by concatenating them
df_split['Volume'] = df_split['Volume1'].astype(str) + df_split['Volume2'].astype(str)
# Check for leading/trailing spaces
df_split['Volume'] = df_split['Volume'].str.strip()
# Then convert to int
df_split['Volume'] = df_split['Volume'].astype(int)
df_split = df_split.drop(columns=['Datetime','Volume1', 'Volume2'])
df_split
```

```python
df_split.info()
```

```python
# Convert data types
df_split['Date'] = pd.to_datetime(df_split['Date'])
df_split.info()
```

```python
#PROBLEM HERE. NEED TO BE A TIME DATA TYPE
pd.to_datetime(df_split['Time'], format='%H:%M:%S').dt.time
# df_split
```

```python
# Optional: Reorder columns
df_split = df_split[['Date', 'Time', 'Open', 'High', 'Low', 'Close', 'Volume']]
```

```python
# Save cleaned data to new CSV (optional)
df_split.to_csv(r'C:\Users\FRED\Desktop\MSE Capstone Project Files\Historical Forex Data\EURUSD_M15(Cleaned).csv', index=False)

# Display first few rows
print(df_split.head())
```
