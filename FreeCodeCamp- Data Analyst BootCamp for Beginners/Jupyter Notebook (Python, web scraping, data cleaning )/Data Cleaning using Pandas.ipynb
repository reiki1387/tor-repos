```python
import pandas as pd
df_list = pd.read_excel(r"C:\Users\FRED\Desktop\FreeCodeCamp- Data Analyst BootCamp for Beginners\PandasYouTubeSeries\Customer Call List.xlsx")
df_list
```

```python
df_list = df_list.drop_duplicates()
df_list
```

```python
df_list = df_list.drop(columns = 'Not_Useful_Column')
df_list
```

```python
df_list["Last_Name"] = df_list["Last_Name"].str.strip("/._")
df_list["Last_Name"]
```

```python
#Capitalize the first letter of the first word ONLY
df_list["Last_Name"] = df_list["Last_Name"].str.capitalize() 
# Capitalize starting letter of each word
df_list["Last_Name"] = df_list["Last_Name"].str.title() 
df_list["Last_Name"]
```

```python
#This will only remove whitespaces in ANY position of the string
df_list["Last_Name"] = df_list["Last_Name"].str.replace(" ", "", regex=False)
```

```python
#This will remove ALL kinds of white spaces like tabs, newlines etc in ANY position of the string
df_list["Last_Name"] = df_list["Last_Name"].str.replace(r"\s+", "", regex=True)
df_list["Last_Name"]
```

```python
df_list
```

```python
#replace ALL with space except ALL that is included in the caret
df_list["Phone_Number"] = df_list["Phone_Number"].str.replace('[^a-zA-Z0-9]', '', regex =True)
df_list["Phone_Number"]
```

```python
#Using regex to insert character in between any data type
df_list['Phone_Number'] = df_list['Phone_Number'].str.replace(r"^(\d{3})(\d{4})(\d+)$", r"\1-\2-\3", regex=True)
df_list['Phone_Number']
```

```python
#replace ALL with space except ALL that is included in the caret
df_list["Phone_Number"] = df_list["Phone_Number"].str.replace('[^a-zA-Z0-9]', '', regex =True)
df_list["Phone_Number"]
```

```python
#Using for loop to insert character in a string datatype
for i in df_list["Phone_Number"].index:
    val = str(df_list.at[i, 'Phone_Number'])  # convert every row to string
    if len(val) >= 7:  # basic check to avoid errors
        formatted = f"{val[:3]}-{val[3:7]}-{val[7:]}"
        df_list.at[i, 'Phone_Number'] = formatted
df_list["Phone_Number"]
```

```python
print(df_list["Phone_Number"].apply(type))
```

```python
#check if there is really a Nan or Null values
df_list['Phone_Number'].isna()
```

```python
#Replace NULL/Nan values with spaces
df_list['Phone_Number'] = df_list['Phone_Number'].fillna('')
df_list['Phone_Number']
```

```python
print(df_list["Phone_Number"].apply(type))
```

```python
df_list['Phone_Number'] = df_list['Phone_Number'].str.replace('Na','', regex = False)
df_list['Phone_Number']
```
